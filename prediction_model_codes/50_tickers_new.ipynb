```python
# Импортируем требуемые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.tsa.stattools import adfuller

# Загрузка исходного файла
input_file = "daily_bars.csv"

# Чтение данных из CSV
df_all = pd.read_csv(input_file)

# Получение списка уникальных тикеров
unique_tickers = df_all['ticker'].unique()

# Ограничение до первых 50 тикеров
limited_tickers = unique_tickers[:50]

# Создание итоговых таблиц для ARIMA и SARIMA
arima_results = []
sarima_results = []

# Цикл обработки по каждому тикеру
for ticker in limited_tickers:
    print(f"Обработка тикера: {ticker}")
    try:
        # Фильтрация данных по тикеру
        df = df_all[df_all['ticker'] == ticker].copy()
        # Загрузка данных для анализа
        df.set_index('price_date', inplace=True)
        df.index = pd.to_datetime(df.index)
        df.rename(columns={
            'price_date': 'Date',
            'close': 'Close_Real',
            'adjusted_close': 'Close'
        }, inplace=True)
        df.index.name = 'Date'
        # Импортируем требуемые библиотеки
        import pandas as pd
        import numpy as np
        from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
        from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
        # Выведим данные
        df.head(10)
        # Выведим размер данных
        print('Размерность датасета: ', df.shape)
        # Выведем информацию о датафрейме
        df.info()
        import matplotlib.pyplot as plt
        # Извлекаем тикер из DataFrame
        ticker = df['ticker'].iloc[0]
        # Строим график временного ряда
        df['Close'].plot(figsize=(15, 6))
        plt.title(f'Цены акций на закрытии ({ticker})')
        plt.xlabel('Дата')
        plt.ylabel('Цена')
        plt.grid()
        plt.show()
        # Проверка пропущенных значений
        print(df.isnull().sum())
        # Вывод строк, содержащих хотя бы одно пустое значение
        print(df[df.isna().any(axis=1)])
        df.interpolate(method='linear', inplace=True)
        # Посмотрим, что получилось
        print(df.isnull().sum())
        # Построим график автокореляции и изучем его показатели
        from pandas.plotting import autocorrelation_plot
        ticker = df['ticker'].iloc[0]
        # Построим график автокорреляции
        plt.figure(figsize=(15, 6))
        autocorrelation_plot(df['Close'])
        plt.title(f'График автокорреляции ({ticker})')
        plt.xlabel('Лаг (задержка)')
        plt.ylabel('Автокорреляция')
        plt.grid(True)
        plt.show()
        # Построим графики графики ACF и PACF
        from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
        ticker = df['ticker'].iloc[0]
        # Строим графики ACF и PACF
        plt.figure(figsize=(12, 6))
        plt.subplot(121)
        plot_acf(df['Close'], ax=plt.gca(), lags=100)
        plt.title(f'ACF (Автокорреляционная функция) для тикера {ticker}')
        plt.subplot(122)
        plot_pacf(df['Close'], ax=plt.gca(), lags=100)
        plt.title(f'PACF (Частичная автокорреляционная функция) для тикера {ticker}')
        plt.tight_layout()
        plt.show()
        # Проверим данные на статичность тестом Дики-Фуллера
        from statsmodels.tsa.stattools import adfuller
        result = adfuller(df['Close'])
        # Выводим результаты теста
        print('Статистика теста Дики-Фуллера:', result[0])
        print('Критические значения:')
        for key, value in result[4].items():
            print(f'\t{key}: {value}')
        # Проведем дифферинцирование
        df['Diff'] = df['Close'].diff()
        # Удалим пустые значения
        df = df.dropna()
        initial_value = df['Close'].iloc[0]
        # Проверим на стационарность
        result = adfuller(df['Diff'])
        # Выводим результаты теста
        print('Статистика теста Дики-Фуллера:', result[0])
        print('Критические значения p-value:', result[1])
        print('Критические значения:')
        for key, value in result[4].items():
            print(f'\t{key}: {value}')

        # Для стационарных данных
        seq_length = 100
        # Размер тренировочной выборки (80%)
        train_size = 0.8
        # Размер тестовой (предсказываемой) выборки
        test_size = int(len(df) * 0.2)
        # Значения для восстановления данных
        last_values_close = df['Close'].iloc[-test_size:]
        last_values_diff = df['Diff'].iloc[-test_size:]
        # Словарь для сохранения данных метрик
        results = {
            'Модель': [],
            'MSE': [],
            'MAE': [],
            'R^2': [],
        }
        # Проверяем, что данные разделились корректно
        print("Последние значения 'Close':")
        print(last_values_close)
        print("Последние значения 'Diff':")
        print(last_values_diff)
        # Общие функции
        # Разделение данных на тренировочные и тестовые
        def split_df(df, seq_len):
            seq_len = int(seq_len)
            split = int(len(df) - seq_len * 2)
            X_train, X_test = df.iloc[:split], df.iloc[split + seq_len:]
            y_train, y_test = df['Diff'].iloc[:split], df['Diff'].iloc[split + seq_len:]
            return X_train, X_test, y_train, y_test

        def create_sequences(data, seq_length):
            sequences = []
            labels = []
            for i in range(0, len(data) - seq_length, seq_length + 1):
                sequences.append(data[i:i + seq_length])
                labels.append(data[i + seq_length])
            return np.array(sequences), np.array(labels)

        # Разделение данных на тренировочные и валидационные RNN, LSTM
        def split_data(data, seq_length, train_size, test_size):
            # Определяем индекс, на котором будет разделение данных
            train_size = int(len(data) * train_size)
            # Разделяем данные на обучающую и валидационную выборки
            train_data = data[:train_size]
            val_data = data[train_size + 2 * seq_length:]
            # Создаем последовательности выборок
            X_train, y_train = create_sequences(train_data, seq_length)
            X_val, y_val = create_sequences(val_data, seq_length)
            # Отделим тестовыю выборку
            split = int(len(X_train) - test_size)
            X_test = X_train[split:]
            y_test = y_train[split:]
            return X_train, y_train, X_val, y_val, X_test, y_test

        # Обратное дифферинцирование
        def revers(predict, last_values):
            pred_rec = np.cumsum(predict) + last_values
            return pred_rec

        # Функция расчета корреляционного коэффициента Пирсона для двух рядов
        def correlate(a, b):
            c = np.corrcoef(a, b)[0, 1]
            return c

        # Функция рисования корреляций прогнозного ряда и исходного со смещением
        def show_corr(y_pred, y_true,
                      back_steps_max=test_size):
            y_len = y_true.shape[0]
            steps = range(0, back_steps_max + 1)
            # Построение графика по всем каналам данных
            plt.figure(figsize=(15, 5))
            # Вычисление коэффициентов корреляции базового ряда и предсказания с разным смещением
            cross_corr = [correlate(y_true[:y_len - step], y_pred[step:]) for step in steps]
            # Вычисление коэффициентов автокорреляции базового ряда с разным смещением
            auto_corr = [correlate(y_true[:y_len - step], y_true[step:]) for step in steps]
            plt.plot(cross_corr, label=f'Прогноз')
            plt.plot(auto_corr, label=f'Истина')
            # Назначение меток шкалы оси x
            plt.xticks(steps)
            plt.xlabel('Шаги смещения')
            plt.ylabel('Коэффициент корреляции')
            plt.title('График автокореляции')
            plt.legend()
            plt.grid(False)
            plt.show()

        # Использование ARIMA и SARIMA
        import pandas as pd
        import matplotlib.pyplot as plt
        import statsmodels.api as sm
        from statsmodels.tsa.arima.model import ARIMA
        # Делаем копию датасета
        df_arima = df.copy()
        # Запишем имя модели
        model_name = 'ARIMA'
        # Сбрасываем индекс с удаленеим столбца
        df_arima = df_arima.reset_index(drop=False)
        # Отделим тренировочную выборку
        X_train_ar, x_test_ar, _, _ = split_df(df_arima, test_size)
        # Читаем параметры из файла
        arima_params = pd.read_csv('arima_params.csv')
        # Получаем уникальный тикер из DataFrame
        current_ticker = df['ticker'].iloc[0]
        # Проверяем, есть ли параметры для текущего тикера
        params = arima_params[arima_params['ticker'] == current_ticker]
        # Устанавливаем параметры ARIMA
        if not params.empty:
            p = int(params['p'].iloc[0])
            d = int(params['d'].iloc[0])
            q = int(params['q'].iloc[0])
        else:
            # Параметры по умолчанию
            p, d, q = 1, 1,
        # Настраиваем и обучаем модель ARIMA
        model_arima = ARIMA(X_train_ar['Diff'], order=(p, d, q))
        model_arima_fit = model_arima.fit()
        # Выводим результат
        print(model_arima_fit.summary())
        print(df_arima.index)
        print(model_arima_fit.data.row_labels)
        # Длина прогноза
        start = len(X_train_ar['Diff']) + test_size
        end = len(df_arima['Diff']) - 1
        # Делаем прогноз, используя обученную модель
        predict = model_arima_fit.predict(start=start, end=end, dynamic=True)
        # Данные для обратного востанавления знчений
        last_values_arima = df_arima['Diff'][-test_size:]
        # Обратное дифферинцирование
        forecast_arima_diff = revers(np.array(predict), last_values_diff)
        # Восстанавливаем оригинальные значения
        forecast_arima_close = revers(np.array(forecast_arima_diff), last_values_close)
        # Преобразуем даты в формат без времени
        df_arima['Date'] = pd.to_datetime(df_arima['Date']).dt.date
        # Извлекаем тикер из DataFrame
        ticker = df_arima['ticker'].iloc[0]
        # Вывод результатов
        plt.figure(figsize=(15, 5))
        plt.plot(df_arima[start:]['Date'], df_arima[start:]['Close'], label=f'Истинные значения ({ticker})',
                 c='darkblue', marker='o')
        plt.plot(df_arima[start:]['Date'], forecast_arima_close, label=f'Прогноз для ({ticker})', c='green',
                 marker='o')
        plt.xlabel('Дата')
        plt.ylabel('Цена')
        plt.title(
            f'ARIMA: Прогноз цен: {ticker} с {df_arima[start:]["Date"].iloc[0]} по {df_arima[start:]["Date"].iloc[-1]}')
        plt.grid()
        plt.legend()
        # Строим графики автокореляции
        show_corr(forecast_arima_close, last_values_close, test_size)
        plt.show()
        # Расчитываем метрики оценки модели
        actual_values_ar = last_values_close
        mse_ar = mean_squared_error(actual_values_ar, forecast_arima_close)
        mae_ar = mean_absolute_error(actual_values_ar, forecast_arima_close)
        r2_ar = r2_score(actual_values_ar, forecast_arima_close)
        # Запишим результаты в таблицу с добавлением тикера
        results['Модель'].append(model_name)
        results['MSE'].append(mse_ar)
        results['MAE'].append(mae_ar)
        results['R^2'].append(r2_ar)
        print(f'\033[1mМетрики оценки модели для тикера {ticker}\033[0m')
        print(f'MSE: {mse_ar}')
        print(f'MAE: {mae_ar}')
        print(f'R^2: {r2_ar}')
        #Модель SARIMA
        import pandas as pd
        import matplotlib.pyplot as plt
        import statsmodels.api as sm
        # Делаем копию датасета
        df_sarima = df.copy()
        # Запишем имя модели
        model_name = 'SARIMA'
        # Разделим данные на выборки
        X_train_sar, x_test_sar, _, _ = split_df(df_sarima, test_size)
        # Читаем параметры SARIMA из файла
        sarima_params = pd.read_csv('sarima_params.csv')
        # Получаем уникальный тикер из DataFrame
        current_ticker = df['ticker'].iloc[0]
        # Проверяем, есть ли параметры для текущего тикера
        sarima_params_for_ticker = sarima_params[sarima_params['ticker'] == current_ticker]
        # Устанавливаем параметры SARIMA
        if not sarima_params_for_ticker.empty:
            sarima_p = int(sarima_params_for_ticker['p'].iloc[0])
            sarima_d = int(sarima_params_for_ticker['d'].iloc[0])
            sarima_q = int(sarima_params_for_ticker['q'].iloc[0])
        else:
            # Параметры по умолчанию для SARIMA
            sarima_p, sarima_d, sarima_q = 5, 0, 1
        # Настраиваем и обучаем модель ARIMA (как пример для SARIMA)
        model_sarima = sm.tsa.statespace.SARIMAX(X_train_sar['Diff'], order=(sarima_p, sarima_d, sarima_q))
        model_fit_sarima = model_sarima.fit()
        # Выводим параметры для проверки
        print(f'Параметры SARIMA для тикера {current_ticker}: p={sarima_p}, d={sarima_d}, q={sarima_q}')
        # Выводим результат
        print(model_fit_sarima.summary())
        # Рассчитываем длину тестового периода
        test_length = len(x_test_sar)
        # Прогноз на весь тестовый период
        forecast_future = model_fit_sarima.get_forecast(steps=test_length)
        # Делаем предсказания и создаем из них Series
        forecast_future = pd.Series(forecast_future.predicted_mean)
        # Создаем даты прогнозов
        future_dates = pd.date_range(start=x_test_sar.index[0], periods=test_length, freq='B')
        # Обратное дифференцирование
        forecast_sarima_diff = revers(np.array(forecast_future), last_values_diff)
        # Восстанавливаем оригинальные значения
        forecast_sarima_close = revers(np.array(forecast_sarima_diff), last_values_close)
        # Создаем новый DataFrame для будущих значений
        forecast_df = pd.DataFrame({'Дата': future_dates, 'Прогноз цены': forecast_sarima_close})
        # Выводим первые строки для проверки
        print(forecast_df.head())
        # Рассчитываем длину тестового периода
        test_length = len(x_test_sar)
        # Прогноз
        forecast_future = model_fit_sarima.get_forecast(steps=test_length)
        # Делаем предсказания и создаем из них Series
        forecast_future = pd.Series(forecast_future.predicted_mean)
        # Создаем даты прогнозов
        future_dates = pd.date_range(start=x_test_sar.index[0], periods=test_length, freq='B')
        # Обратное дифферинцирование
        forecast_sarima_diff = revers(np.array(forecast_future), last_values_diff)
        # Восстанавливаем оригинальные значения
        forecast_sarima_close = revers(np.array(forecast_sarima_diff), last_values_close)
        # Создаем новый DataFrame для будущих значений
        forecast_df = pd.DataFrame({'Дата': future_dates, 'Прогноз цены': forecast_sarima_close})
        # Извлекаем тикер из DataFrame
        ticker = df['ticker'].iloc[0]
        # Визуализация исходных данных и прогноза
        plt.figure(figsize=(15, 5))
        plt.plot(last_values_close, label=f'Истинные значения ({ticker})', c='darkblue', marker='o')
        plt.plot(future_dates, forecast_df['Прогноз цены'], label=f'Предсказанные значения ({ticker})', c='green',
                 marker='o')
        plt.title(
            f'SARIMA: Прогноз цен: {ticker} с {str(future_dates[0]).split()[0]} по {str(future_dates[-1]).split()[0]}')
        plt.xlabel('Дата')
        plt.ylabel('Цена')
        plt.legend()
        plt.grid(True)
        # Строим графики автокореляции
        show_corr(forecast_df['Прогноз цены'], last_values_close, test_size)
        plt.show()
        # Разсчитываем метрики оценки модели
        actual_values_sar = last_values_close
        mse_sar = mean_squared_error(actual_values_sar, forecast_df[-test_size:]['Прогноз цены'])
        mae_sar = mean_absolute_error(actual_values_sar, forecast_df[-test_size:]['Прогноз цены'])
        r2_sar = r2_score(actual_values_sar, forecast_df[-test_size:]['Прогноз цены'])
        # Запишим результаты в таблицу с добавлением тикера
        results['Модель'].append(model_name)
        results['MSE'].append(mse_sar)
        results['MAE'].append(mae_sar)
        results['R^2'].append(r2_sar)
        print(f'\033[1mМетрики оценки модели для тикера {ticker}\033[0m')
        print(f'MSE: {mse_sar}')
        print(f'MAE: {mae_sar}')
        print(f'R^2: {r2_sar}')
        import os
        import pandas as pd
        # Получаем тикер
        ticker = df['ticker'].iloc[0]
        # Создаем DataFrame для записи новых данных
        forecast_data = pd.DataFrame({
            'Дата': forecast_df['Дата'],
            'Прогноз цены': forecast_df['Прогноз цены'],
            'Фактическая цена': last_values_close,
            'Тикер': [ticker] * len(forecast_df['Дата'])
        })
        # Путь к файлу
        file_path = "all_forecasts_data.csv"
        # Проверяем, существует ли файл
        if os.path.exists(file_path):
            # Если файл существует, читаем его
            existing_data = pd.read_csv(file_path)
            # Добавляем новые данные в существующий DataFrame
            combined_data = pd.concat([existing_data, forecast_data], ignore_index=True)
        else:
            # Если файла нет, используем только новые данные
            combined_data = forecast_data
        # Сохраняем итоговый DataFrame в файл
        combined_data.to_csv(file_path, index=False, encoding='utf-8-sig')
        print(f"Данные сохранены в файл {file_path}")
        print(f'\033[1mМетрики оценки модели для тикера {ticker}\033[0m')
        print(f'MSE: {mse_sar}')
        print(f'MAE: {mae_sar}')
        print(f'R^2: {r2_sar}')
    except np.linalg.LinAlgError:
        print(f'Ошибка LinAlgError для тикера {current_ticker}. Тикер пропущен.')
        continue
    except Exception as e:
        print(f'Неизвестная ошибка для тикера {current_ticker}: {e}. Тикер пропущен.')
        continue
```
